```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pyreadr

result = pyreadr.read_r('DB_EXTRACT_CLOUD.RData')
```

```python
result.keys()
```

```python
raw_pipe = result['SEND_TO_1']
raw_copa = result['SEND_TO_2']
raw_copa.columns = ['YEAR_QUARTER', 'REGION',  'ACTUALS']
```

```python
raw_pipe['DATE'] = raw_pipe[['DATE']].astype(str)
raw_pipe['REGION'] = raw_pipe[['REGION']].astype(str)
raw_pipe['REGION2'] = raw_pipe[['REGION2']].astype(str)
raw_pipe['YEAR_QUARTER'] = raw_pipe[['YEAR_QUARTER']].astype(str)

raw_copa['YEAR_QUARTER'] = raw_copa[['YEAR_QUARTER']].astype(str)
raw_copa['REGION'] = raw_copa[['REGION']].astype(str)
```

```python
raw_pipe.columns
```

```python
raw_pipe.loc[(raw_pipe.REGION=='nan'),'REGION'] = 'NOAM'

raw_copa['REGION'] = raw_copa['REGION'].str.replace('ap','AP')
raw_copa['REGION'] = raw_copa['REGION'].str.replace('emnorth','EMNORTH')
raw_copa['REGION'] = raw_copa['REGION'].str.replace('emsouth','EMSOUTH')
raw_copa['REGION'] = raw_copa['REGION'].str.replace('gcn','GCN')
raw_copa['REGION'] = raw_copa['REGION'].str.replace('la','LA')
raw_copa['REGION'] = raw_copa['REGION'].str.replace('mee','MEE')
raw_copa['REGION'] = raw_copa['REGION'].str.replace('na','NOAM')

raw_pipe = raw_pipe[(raw_pipe.REGION!='GEOBND')]
raw_copa = raw_copa[(raw_copa.REGION=='AP') | (raw_copa.REGION=='EMNORTH') | 
                   (raw_copa.REGION=='EMSOUTH') | (raw_copa.REGION=='GCN') |
                   (raw_copa.REGION=='LA') | (raw_copa.REGION=='MEE') |
                   (raw_copa.REGION=='NOAM')]
                                        
```

```python
print(raw_pipe.REGION.unique())
print(raw_copa.REGION.unique())
```

```python
df = pd.merge(raw_pipe, raw_copa,
             on=['YEAR_QUARTER', 'REGION'],
             how='left')
```

```python
df.to_csv('./df_cloud.csv', sep = ';', decimal=',')
```
